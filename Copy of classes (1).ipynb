```python
class Crop:
    def __init__(self, name, season, expected_yield):
        self.name = name
        self.season = season
        self.expected_yield = expected_yield

rice = Crop("rice", "kharif", 5.2)
wheat = Crop("wheat", "rabi", 4.1)
print(rice.name, rice.season, rice.expected_yield)
print(wheat.name, wheat.season, wheat.expected_yield)
```

```python
class Crop:
  def __init__(self, name, season, expected_yield):
    self.name = name
    self.season = season
    self.expected_yield = expected_yield

  def total_production(self,area):
    return self.expected_yield * area

area=10
rice=Crop("rice","kharif",5.2)
print("total rice production:",rice.total_production(area),"tons")
```

```python
rice = Crop("rice","kharif",5.2)
wheat = Crop("wheat","rabi",4.1)
corn = Crop("corn","kharif",6.8)
cotton = Crop("cotton","kharif",2.3)

crops_list = [rice,wheat,corn,cotton]

for crop_obj in crops_list:
    production = crop_obj.total_production(10)
    print(f"{crop_obj.name}: {production} tons")
```
